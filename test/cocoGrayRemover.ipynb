```python
import torch
import os
import torchvision
DATAPATH = '/mnt/NeuralNetworksDL/coco/'
```

```python
filenames = os.listdir(DATAPATH)
for file in filenames:
    img = torchvision.io.read_image(DATAPATH + file).to(torch.float32)
    if img.shape[0] == 1:
        os.system('rm ' + DATAPATH + file)
        print(img.shape, 'deleted' + file)
```
